```python
import numpy as np
import matplotlib.pyplot as plt

# Przykładowe dane - rzeczywiste i przewidywane wartości
y_true = np.array([2, 4, 5, 4, 5])
y_pred = np.array([1.5, 3.5, 4.5, 4, 5.5])

# Funkcja obliczająca L2 Loss
def l2_loss(y_true, y_pred):
    return np.sum((y_true - y_pred)**2)

# Wygenerowanie punktów na wykresie
shifts = np.arange(-10, 11, 0.1)
loss_values = [l2_loss(y_true, y_pred + shift) for shift in shifts]

# Wykres
plt.plot(shifts, loss_values)
plt.title('L2 Loss')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cubed_hinge_loss(y_true, y_pred):
    loss = np.maximum(0, 1 - y_true * y_pred)**3
    return np.mean(loss)

# Przykładowe dane - rzeczywiste i przewidywane wartości
y_true = np.array([1, -1, 1, -1, 1])
y_pred_range = np.linspace(-2, 2, 100)  # Zakres przewidywanych wartości

# Wygenerowanie wartości straty dla różnych przewidywanych wartości
loss_values = [cubed_hinge_loss(y_true, y_pred) for y_pred in y_pred_range]

# Wykres
plt.plot(y_pred_range, loss_values)
plt.title('Cubed Hinge Loss')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cross_entropy_loss(y_true, y_pred):
    epsilon = 1e-15  # A small constant to avoid log(0)
    loss = - (y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))
    return np.mean(loss)

# Przykładowe dane - rzeczywiste etykiety (0 lub 1) i przewidywane prawdopodobieństwa
y_true = np.array([1, 0, 1, 0, 1])
y_pred_range = np.linspace(0.01, 0.99, 100)  # Zakres przewidywanych prawdopodobieństw

# Wygenerowanie wartości straty dla różnych przewidywanych prawdopodobieństw
loss_values = [cross_entropy_loss(y_true, y_pred) for y_pred in y_pred_range]

# Wykres
plt.plot(y_pred_range, loss_values)
plt.title('Cross Entropy Loss')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generowanie danych
x = np.linspace(-6, 6, 100)  # Zakres danych od -6 do 6
y = sigmoid(x)

# Rysowanie funkcji sigmoidalnej
plt.plot(x, y, label='Sigmoid')
plt.title('Sigmoid')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

# Generowanie danych
x = np.linspace(-2, 2, 100)
y_relu = relu(x)

# Rysowanie funkcji ReLU
plt.plot(x, y_relu, label='ReLU')
plt.title('ReLU')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.02):
    return np.where(x > 0, x, alpha * x)

# Generowanie danych
x = np.linspace(-2, 2, 100)
y_leaky_relu = leaky_relu(x)

# Rysowanie funkcji Leaky ReLU
plt.plot(x, y_leaky_relu, label='Leaky ReLU (alpha=0.02)')
plt.title('Leaky ReLU')
plt.show()
```
