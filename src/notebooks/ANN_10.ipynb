```python
import pandas as pd

df = pd.read_csv('../../output_data/4_2_vectorized_ds_glove_25.csv')

df.info()
```

```python
df.head()
```

```python
df = df.fillna(0)

df.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(['tag', 'Unnamed: 0'], axis=1)
y = df['tag']

y = pd.get_dummies(y)

y = y.astype('float32')

y.shape
```

```python
x_shape = X.shape[1]

x_shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print(X_train.info())
print(X_test.info())
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(1100, activation='relu'))
model.add(tf.keras.layers.Dense(4*1100, activation='relu'))
model.add(tf.keras.layers.Dense(8*1100, activation='relu'))
model.add(tf.keras.layers.Dense(8*1100, activation='relu'))
model.add(tf.keras.layers.Dense(8*1100, activation='relu'))
model.add(tf.keras.layers.Dense(4*1100, activation='relu'))
model.add(tf.keras.layers.Dense(2*1100, activation='relu'))
model.add(tf.keras.layers.Dense(31, activation='softmax'))

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0006),
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=['accuracy'] 
    )
```

```python
X_train_np = X_train.to_numpy()
y_train_np = y_train.to_numpy()
X_test_np = X_test.to_numpy()
y_test_np = y_test.to_numpy()

model.fit(X_train_np, y_train_np, validation_data=(X_test_np, y_test_np), batch_size=256, epochs=25)
```

```python
X_test_np = X_test.to_numpy()
y_pred = np.argmax(model.predict(X_test_np), axis=1)


y_test_labels = np.argmax(y_test.values, axis=1)

accuracy = accuracy_score(y_test_labels, y_pred)
print(f'Accuracy: {accuracy}')
```
